# Mathematics Example: Symbolic Calculus

We will use SymPy to differentiate and integrate functions.

```python
from sympy import symbols, diff, integrate, sin, exp

x = symbols('x')
f = exp(-x) * sin(x)

# Derivative
df = diff(f, x)
print(f"Derivative: {df}")

# Indefinite Integral
int_f = integrate(f, x)
print(f"Integral: {int_f}")
```
