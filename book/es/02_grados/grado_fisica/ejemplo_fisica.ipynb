# Ejemplo de Física: Oscilador Armónico

Este notebook simula un oscilador armónico simple.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
k = 1.0  # Constante del resorte
m = 1.0  # Masa
omega = np.sqrt(k / m)
t = np.linspace(0, 20, 100)

# Posición
x = np.cos(omega * t)

# Gráfico
plt.figure(figsize=(8, 4))
plt.plot(t, x)
plt.title('Oscilador Armónico Simple')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.grid(True)
plt.show()
```
