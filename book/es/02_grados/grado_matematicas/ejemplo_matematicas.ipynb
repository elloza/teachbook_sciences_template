# Ejemplo de Matemáticas: Cálculo Simbólico

Vamos a usar SymPy para derivar e integrar funciones.

```python
from sympy import symbols, diff, integrate, sin, exp

x = symbols('x')
f = exp(-x) * sin(x)

# Derivada
df = diff(f, x)
print(f"Derivada: {df}")

# Integral indefinida
int_f = integrate(f, x)
print(f"Integral: {int_f}")
```
